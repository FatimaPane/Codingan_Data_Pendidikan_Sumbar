```python
# Import libraries
import pandas as pd
import folium
from geopy.geocoders import Nominatim
from time import sleep

# Load data
df = pd.read_excel("data_pendidikan_sumbar_final.xlsx")

# Pastikan data numerik
df_numeric = df.drop(columns=['Kabupaten_Kota'])
df_numeric = df_numeric.apply(pd.to_numeric, errors='coerce')

# Buat indikator rata-rata per kabupaten
df['Rata_Rata'] = df_numeric.mean(axis=1)

# Perbaiki nama kabupaten
def fix_nama_kota(nama):
    nama = nama.replace("Kab.", "Kabupaten")
    nama = nama.replace("Kota", "Kota")  # tetap (optional)
    return nama.strip()
    
df['Nama_Geo'] = df['Kabupaten_Kota'].apply(fix_nama_kota)

# Ambil daftar kabupaten
unique_cities = df['Nama_Geo'].unique()

# Inisialisasi geocoder
geolocator = Nominatim(user_agent="geoapi")

# Ambil koordinat
city_coords = {}

for city in unique_cities:
    try:
        location = geolocator.geocode(city + ", Sumatera Barat, Indonesia")
        if location:
            city_coords[city] = (location.latitude, location.longitude)
        else:
            print(f"Koordinat untuk {city} tidak ditemukan.")
        sleep(1)
    except Exception as e:
        print(f"Error geocoding {city}: {e}")

# Inisialisasi peta (Sumatera Barat)
m = folium.Map(location=[-0.95, 100.35], zoom_start=7)

# Tambahkan marker
for city, coords in city_coords.items():
    nilai = df[df['Nama_Geo'] == city]['Rata_Rata'].values[0]

    folium.Marker(
        location=coords,
        popup=f"{city}<br>Rata-rata: {nilai:.2f}",
        tooltip=city,
        icon=folium.Icon(color="green" if nilai > df['Rata_Rata'].mean() else "red")
    ).add_to(m)

# Simpan peta
m.save("peta_pendidikan_sumbar.html")
m
```
